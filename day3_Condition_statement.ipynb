```python
# if-elif-else

print("Write the  positive numbers only")
n = int(input("Enter your number here:"))

if (n==0 and n<=10):
  print("the value of n is between the range 1 - 10\n")

elif (n>=11 and n<=20):
  print("the value of n is between 11-20\n")

else:
  print("you write the code greater than 20\n")



```

```python
marks = int(input("Enter marks: "))

if marks < 0 or marks > 100:
    print("Invalid marks")
elif marks >= 95:
    print("A Grade")
elif marks >= 90:
    print("B Grade")
elif marks >= 75:
    print("C Grade")
elif marks >= 60:
    print("D Grade")
else:
    print("FAIL")
```
