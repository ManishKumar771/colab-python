```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna(inplace=True)
print(df)
```

```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna()
print(df)

# fillna | null -> fill
df.fillna(5)
```

```python
# sub plot
import matplotlib.pyplot as plt #

# pie chart data
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
plt.subplot(1,2,1)
plt.pie(y,labels=x)

# line chart data
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']
plt.subplot(1,2,2)
plt.plot(x,y)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['apple', 'banana', 'carrot', 'mango']
values = [30, 50, 70, 5]

plt.style.use('seaborn-v0_8-whitegrid')

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Alternative Data Visualizations', fontsize=16, fontweight='bold', y=1.02)

axs[0, 0].pie(values, labels=categories, startangle=140,
              wedgeprops=dict(width=0.4, edgecolor='w'))
axs[0, 0].set_title('Donut Chart')

axs[0, 1].stem(categories, values, basefmt=" ")
axs[0, 1].set_ylim(0, 80)
axs[0, 1].set_title('Lollipop Plot')

axs[1, 0].fill_between(categories, values, color="skyblue", alpha=0.5)
axs[1, 0].plot(categories, values, color="dodgerblue", linewidth=2)
axs[1, 0].set_ylim(0, 80)
axs[1, 0].set_title('Area Chart')

sizes = [v * 15 for v in values]
colors = values
scatter = axs[1, 1].scatter(categories, values, s=sizes, c=colors,
                            cmap='plasma', alpha=0.7, edgecolors='w', linewidth=2)
axs[1, 1].set_ylim(0, 80)
axs[1, 1].set_title('Bubble Chart')

plt.tight_layout()

plt.show()
```
